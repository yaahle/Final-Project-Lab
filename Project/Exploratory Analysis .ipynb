```python
import csv
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
import math

data_df_1 = pd.read_csv("loans.csv")
data_df_2 = pd.read_csv("loan_test.csv")
data_df_2['Status'] = ''

data_df = data_df_1.append(data_df_2, ignore_index=True)
```

```python
Gender = data_df['Gender']
Gender_nums = np.array([np.count_nonzero(Gender == "Male"), np.count_nonzero(Gender == "Female"), 
                        len(Gender) - (np.count_nonzero(Gender == "Male") + np.count_nonzero(Gender == "Female"))])

Genderlabels = ["Male {n:.2f}%".format(n = 100*Gender_nums[0]/len(Gender)), "Female {n:.2f}%".format(n = 100*Gender_nums[1]/len(Gender)), "NA {n:.2f}%".format(n = 100*Gender_nums[2]/len(Gender))]
Gendercolors = ["#ADD8E6", "#e6ade0", "#000000"]

plt.pie(Gender_nums, labels = Genderlabels, colors = Gendercolors)
plt.show()
```

```python
Married = data_df['Married']
Married_nums = np.array([np.count_nonzero(Married == "Yes"), np.count_nonzero(Married == "No"), 
                        len(Married) - (np.count_nonzero(Married == "Yes") + np.count_nonzero(Married == "No"))])

Marriedlabels = ["Yes {n:.2f}%".format(n = 100*Married_nums[0]/len(Married)), "No {n:.2f}%".format(n = 100*Married_nums[1]/len(Married)),
                 "NA {n:.2f}%".format(n = 100*Married_nums[2]/len(Married))]
Marriedcolors = ["#29FF4D", "#E01D45", "#000000"]

plt.pie(Married_nums, labels = Marriedlabels, colors = Marriedcolors)
plt.show()
```

```python
Dependents = data_df['Dependents']
Dependents_nums = np.array([np.count_nonzero(Dependents == "0"), np.count_nonzero(Dependents == "1"), 
                            np.count_nonzero(Dependents == "2"), np.count_nonzero(Dependents == "3+"),  
                        len(Dependents) - (np.count_nonzero(Dependents == "0") + np.count_nonzero(Dependents == "1")
                                          + np.count_nonzero(Dependents == "2") + np.count_nonzero(Dependents == "3+"))])

Dependentslabels = ["Zero {n:.2f}%".format(n = 100*Dependents_nums[0]/len(Dependents)),
                    "One {n:.2f}%".format(n = 100*Dependents_nums[1]/len(Dependents)),
                    "Two {n:.2f}%".format(n = 100*Dependents_nums[2]/len(Dependents)),
                    "Three+ {n:.2f}%".format(n = 100*Dependents_nums[3]/len(Dependents)),
                    "NA {n:.2f}%".format(n = 100*Dependents_nums[4]/len(Dependents))]
Dependentscolors = ["#EAED0C", "#F7CF0C", "#E19100", "#F7700C", "#000000"]

plt.pie(Dependents_nums, labels = Dependentslabels, colors = Dependentscolors)
plt.show()
```

```python
Education = data_df['Education']
Education_nums = np.array([np.count_nonzero(Education == "Graduate"), np.count_nonzero(Education == "Not Graduate"), 
                        len(Education) - (np.count_nonzero(Education == "Graduate") + np.count_nonzero(Education == "Not Graduate"))])

Educationlabels = ["Graduate {n:.2f}%".format(n = 100*Education_nums[0]/len(Education)), "Not Graduate {n:.2f}%".format(n = 100*Education_nums[1]/len(Education)), "NA {n:.2f}%".format(n = 100*Education_nums[2]/len(Education))]
Educationcolors = ["#5300EB", "#FFF300", "#000000"]

plt.pie(Education_nums, labels = Educationlabels, colors = Educationcolors)
plt.show()
```

```python
Employment = data_df['Self_Employed']
Employment_nums = np.array([np.count_nonzero(Employment == "Yes"), np.count_nonzero(Employment == "No"), 
                        len(Employment) - (np.count_nonzero(Employment == "Yes") + np.count_nonzero(Employment == "No"))])

Employmentlabels = ["Self Employed {n:.2f}%".format(n = 100*Employment_nums[0]/len(Employment)), "Not Self Employed {n:.2f}%".format(n = 100*Employment_nums[1]/len(Employment)), "NA {n:.2f}%".format(n = 100*Employment_nums[2]/len(Employment))]
Employmentcolors = ["#92FA00", "#D519FF", "#000000"]

plt.pie(Employment_nums, labels = Employmentlabels, colors = Employmentcolors)
plt.show()
```

```python
Income = data_df['Applicant_Income']
print("Number of missing elements: {}".format(len(Income) - np.count_nonzero(Income)))
plt.hist(Income, bins = list(range(1, 9000000, 200000)))
plt.show()

plt.boxplot(Income)
plt.show()
```

```python
CoIncome = data_df['Coapplicant_Income']
print("Number of missing elements: {}".format(len(CoIncome) - np.count_nonzero(CoIncome)))
plt.hist(CoIncome, bins = list(range(1, 4500000, 100000)))
plt.show()

plt.boxplot(CoIncome)
plt.show()
```

```python
Loans = data_df['Loan_Amount']
print("Number of missing elements: {}".format(len(Loans) - np.count_nonzero(Loans)))
plt.hist(Loans, bins = list(range(1, 70000000, 1000000)))
plt.show()

plt.boxplot(Loans)
plt.show()
```

```python
Term = data_df['Term'] / 12
print("Number of missing elements: {}".format(np.count_nonzero([math.isnan(e) for e in Term])))
#Term[math.isnan(Term)] = -1
plt.hist(Term, bins = list(range(-5, 45, 5)))
plt.show()
```

```python
Credit = data_df['Credit_History']
Credit_nums = np.array([np.count_nonzero(Credit == 1), np.count_nonzero(Credit == 0), 
                        len(Credit) - (np.count_nonzero(Credit == 1) + np.count_nonzero(Credit == 0))])

Creditlabels = ["Yes {n:.2f}%".format(n = 100*Credit_nums[0]/len(Credit)), "No {n:.2f}%".format(n = 100*Credit_nums[1]/len(Credit)),
                 "NA {n:.2f}%".format(n = 100*Credit_nums[2]/len(Credit))]
Creditcolors = ["#29FF4D", "#E01D45", "#000000"]

plt.pie(Credit_nums, labels = Creditlabels, colors = Creditcolors)
plt.show()
```

```python
Area = data_df['Area']
Area_nums = np.array([np.count_nonzero(Area == "Semiurban"), np.count_nonzero(Area == "Urban"), np.count_nonzero(Area == "Rural")])

Arealabels = ["Semiurban {n:.2f}%".format(n = 100*Area_nums[0]/len(Area)), "Urban {n:.2f}%".format(n = 100*Area_nums[1]/len(Area)),
                 "Rural {n:.2f}%".format(n = 100*Area_nums[2]/len(Area))]
Areacolors = ["#190CF6", "#8C00EB", "#F60CB8"]

plt.pie(Area_nums, labels = Arealabels, colors = Areacolors)
plt.show()
```
